```python
import pandas as pd
```

```python
path = "diabetes_prediction_dataset.csv"
diabetes_df=pd.read_csv(path)
```

```python
diabetes_df
```

```python
diabetes_df.info()
```

```python
bins = [0,12,18,35,65,80]
labels= ['children','teen','young_adults','adults','senior']
diabetes_df['binned_age'] = pd.cut(diabetes_df['age'],bins=bins,labels=labels)
diabetes_df
```

```python
diabetes_df['smoking_history'].value_counts()
```

```python
diabetes_df= diabetes_df[diabetes_df!= 'No Info'].dropna()
diabetes_df['smoking_history'].value_counts()
```

```python
diabetes_df['smoking_history'] = diabetes_df['smoking_history'].replace('ever', 'never')
diabetes_df['smoking_history'].value_counts()
```

```python
diabetes_df.to_csv('diabetes_clean_df', index=False)
```
